```python
# Cell 2: Import the Matplotlib plotting module
# This line imports the plotting module from the matplotlib library.
# We give it the alias 'plt' as a standard convention for easier use.
import matplotlib.pyplot as plt
```

```python
# Define the data points for the x-axis and y-axis.
# In financial analysis, x could be a timeline (like months),
# and y could be a financial metric (like a price or portfolio value).
x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 40, 50]

# This is the main command to generate a line chart. It takes the x and y
# data points and plots them on a graph.
plt.plot(x, y)

# This command displays the plot in your output.
plt.show()
```

```python
# Define the data points for the x-axis and y-axis.
# In financial analysis, 'x' could be a timeline (like quarters or years) and 'y' could be a financial metric,
# such as a stock's price or a company's revenue.
x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 40, 50]

# plt.plot(x, y) draws the line using your data points.
plt.plot(x, y)

# plt.xlabel("X axis") adds a clear and descriptive label to the horizontal axis.
# This is crucial for making your financial charts easy to read and understand.
plt.xlabel("X axis")

# plt.ylabel("Y axis") adds a label to the vertical axis, explaining what the values represent.
plt.ylabel("Y axis")

# plt.title("Plot test") gives your chart a title, which is a key part of any financial visualization.
# A clear title helps explain the chart's purpose at a glance.
plt.title("Plot test")

# plt.show() displays the complete chart with all the labels and the title.
plt.show()
```

```python
# plt.bar(x, y) creates the bar chart. It takes the x and y data points and draws a bar for each pair.
plt.bar(x, y)

# plt.show() displays the chart in your output.
plt.show()
```

```python
# plt.scatter(x, y) creates the scatter plot. It takes the x and y data points and plots each pair as a dot.
# This helps you visually identify if there is a trend or a correlation between the two variables.
plt.scatter(x, y)

# plt.show() displays the chart in your output.
plt.show()
```

```python
import pandas as pd
df = pd.read_csv('house_sales_in_king_county_usa.csv')
```

```python
# Create a histogram to visualize the distribution of house prices.
# 'df["price"]' selects the column with all the house prices.
# 'bins=100' tells the plot to create 100 "buckets" or intervals to group the prices.
# This helps you see how often houses fall into a specific price range.
plt.hist(df["price"], bins=100)

# The 'plt.show()' command displays the histogram.
plt.show()
```

```python
# This line of code is a two-step process:
# 1. First, we filter the DataFrame. The part `df.loc[df["price"] < 2000000]`
#    selects all the rows where the house price is less than $2,000,000.
#    This is like an audit: you are only looking at the transactions that fall within a specific range.
# 2. Then, from that filtered data, we select just the "price" column.
#
# This allows us to focus the histogram on the majority of sales and avoid having
# a few very expensive outliers skew the visual representation.
plt.hist(df.loc[df["price"] < 2000000]["price"], bins=100)

# The 'plt.show()' command displays the histogram.
plt.show()
```

```python
# The first crucial step: convert the 'date' column from a simple string to a datetime object.
# This allows pandas to understand and perform time-based operations, like resampling.
df['date'] = pd.to_datetime(df['date'])

# This is the core of the analysis. We resample the data to a monthly frequency.
# '.resample('M', on='date')' groups the data by month ('M').
# '.mean()' then calculates the average of all columns for each month.
# This helps to smooth out the data and reveal the overall trend in house prices.
df_grouped = df.resample('M', on='date').mean()

# Now we plot the resampled data. The x-axis is the index of the grouped DataFrame (the months),
# and the y-axis is the average price for each month.
plt.plot(df_grouped.index, df_grouped['price'])

# Add clear and descriptive labels to the chart. This is essential for professional reporting.
plt.xlabel('Date')
plt.ylabel('Average Price')
plt.title('Trend of Average House Prices Over Time')

# Display the final chart.
plt.show()
```
